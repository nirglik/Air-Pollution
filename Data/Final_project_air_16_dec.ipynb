```python
# install if needed
import json
import requests
import pandas as pd 
```

```python
#read the total data (file that contains all the data "pushed" from the Luftdaten Website using API). 
#It will load the csv file and transforn in dataframe, so that we can add the new data (daily). 

df_lqi_total = pd.read_csv("data/df_lqi_total.csv",index_col=0)
df_pm10_total = pd.read_csv("data/df_pm10_total.csv",index_col=0)
df_no2_total= pd.read_csv ("data/df_no2_total.csv",index_col=0)
df_o3_total = pd.read_csv ("data/df_o3_total.csv",index_col=0)
df_co_total = pd.read_csv("data/df_co_total.csv",index_col=0)
```

```python
df_no2_total.shape
```

```python
response = requests.get("https://luftdaten.berlin.de/api/lqis/data")
print(response.status_code)
results = response.json()
```

```python
results
```

```python
len(results)
```

```python
pd.json_normalize(response.json())
```

```python
df = pd.DataFrame(response.json()) 
```

```python
df['data']
```

```python
stations=[]
for i in response.json():
    stations.append(i['data'])
```

```python
len(stations)
```

```python
stations
```

```python
stations_list=[]
for i in stations:
    stations_list.extend(i)
stations_list
```

```python
df_air=pd.json_normalize(stations_list)
```

```python
df_air.loc[df_air['component']=='lqi'] #separating the data according to the components (lqi, no2, ...)
```

```python
df_lqi_day = df_air.loc[df_air['component']=='lqi'] #daily df per component 
df_pm10_day = df_air.loc[df_air['component']=='pm10']
df_no2_day = df_air.loc[df_air['component']=='no2']
df_o3_day = df_air.loc[df_air['component']=='o3']
df_co_day = df_air.loc[df_air['component']=='co']
```

```python
df_lqi_total
```

```python
df_lqi_total.info()
```

```python
#concatanating total data 

df_lqi_total = pd.concat([df_lqi_total, df_lqi_day], axis=0)
df_lqi_total = df_lqi_total.reset_index(drop=True) #DROP=TRUE, does not adds a index column

df_pm10_total = pd.concat([df_pm10_total, df_pm10_day], axis=0)
df_pm10_total = df_pm10_total.reset_index(drop=True)

df_no2_total = pd.concat([df_no2_total, df_no2_day], axis=0)
df_no2_total = df_no2_total.reset_index(drop=True)

df_o3_total = pd.concat([df_o3_total, df_o3_day], axis=0)
df_o3_total = df_o3_total.reset_index(drop=True)

df_co_total = pd.concat([df_co_total, df_co_day], axis=0)
df_co_total = df_co_total.reset_index(drop=True)
```

```python
#Saving the concataneted data

df_lqi_total = df_lqi_total.to_csv("data/df_lqi_total.csv")
df_pm10_total=df_pm10_total.to_csv("data/df_pm10_total.csv")
df_no2_total=df_no2_total.to_csv("data/df_no2_total.csv")
df_o3_total=df_o3_total.to_csv("data/df_o3_total.csv")
df_co_total=df_co_total.to_csv("data/df_co_total.csv")
```

```python
#Saving the daily data

df_lqi_day = df_lqi_day.to_csv("data/df_lqi_day.csv")
df_pm10_day = df_pm10_day.to_csv("data/df_pm10_day.csv")
df_no2_day = df_no2_day.to_csv("data/df_no2_day.csv")
df_o3_day = df_o3_day.to_csv("data/df_o3_day.csv")
df_co_day = df_co_day.to_csv("data/df_co_day.csv")
```
